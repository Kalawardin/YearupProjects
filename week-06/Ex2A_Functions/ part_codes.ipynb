```python
def get_suplier_code(code):
    return code.split(':')[0]
def get_product_num(code):
    return code.split(':')[1].split('-')[0]
def get_size(code):
    return code.split('-')[1]
```

```python
code1 = "ACME:123-L"
code2 = "DI:12345-M"
code3 = "ACE:1-12"
```

```python
print(f"For {code1}, the supplier is {get_suplier_code(code1)}. Part number {get_product_num(code1)} is available in size {get_size(code1)}.")
print(f"For {code2}, the supplier is {get_suplier_code(code2)}. Part number {get_product_num(code2)} is available in size {get_size(code2)}.")
print(f"For {code3}, the supplier is {get_suplier_code(code3)}. Part number {get_product_num(code3)} is available in size {get_size(code3)}.")
```
