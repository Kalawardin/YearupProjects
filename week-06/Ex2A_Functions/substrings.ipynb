```python
## Lab 1. Working with names – parsing strings
```

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i+1:]
            print(f"Full name: {name}")
            print(f"First name: {first}")
            print(f"Last name: {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen','Ken Penn']
```

```python
parse_and_display(test_list[0])
```

```python
parse_and_display('Fuat Baslar')
```

```python
def parse_and_display2(name):
    spaces = []
    for i in range(len(name)):
        if name[i] == ' ':
            spaces.append(i)
    
    if len(spaces) == 0:
        print(f"Full name: {name}")
        print(f"first name: {name}")
    elif len(spaces) == 1:
        print(f"Full name: {name}")
        first = name[:spaces[0]]
        last = name[spaces[0]+1:]
        print(f"First name: {first}")
        print(f"Last name : {last}")
    elif len (spaces) == 2:
        first = name[:spaces[0]]
        middle = name[spaces[0]+1:spaces[-1]]
        last = name[spaces[-1]+1:]
        print(f"Full name: {name}")
        print(f"First name: {first}")
        print(f"Middle name: {middle}")
        print(f"Last name: {last}")
    else:
        print("Please try again. Name is invalid.")
```

```python
parse_and_display2('Megan Thee Stallion')
```
