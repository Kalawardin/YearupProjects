```python
import datetime
```

```python
# Today
today = datetime.datetime.now()
```

```python
print(today.year)
print(today.strftime("%A"))
print(today.strftime("%B"))
print(today.strftime("%D"))
```

```python
# My birthday
```

```python
birthday = datetime.date(2001 , 7 , 23)

print(birthday)
```

```python
print(f"My birthday is {birthday.strftime("%B")}/{birthday.strftime("%d")}/{birthday.year}")
```

```python
# Next 90 Days
```

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f"90 days from today is {ninety_d.strftime("%B")} {ninety_d.strftime("%d")}, {ninety_d.year}")
```

```python
# Time for Dinner
```

```python
dinner_time = datetime.time()
```

```python
print(f"Let’s meet for dinner at {dinner_time.strftime("%H")}:{dinner_time.strftime("%M")} {dinner_time.strftime("%p")}")
```
