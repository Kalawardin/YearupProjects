```python
def get_soc_sec_tax(gross_pay):
    ss_tax = gross_pay * 0.062 
    return ss_tax

def get_medicare_tax(gross_pay):
    med_tax = gross_pay * 0.0145
    return med_tax

def get_federal_tax(gross_pay,withholding_code):
    
    if withholding_code == 0:
        fed_tax = gross_pay * 0.23
    elif withholding_code == 1:
        fed_tax = gross_pay * 0.21
    elif withholding_code == 2:
        fed_tax = gross_pay * 0.195
    elif withholding_code == 3:
        fed_tax = gross_pay * 0.185
    elif withholding_code > 3:
        fed_tax = gross_pay * (0.18 - ((withholding_code - 4) * 0.005))
    else:
        print("Please enter valid code!")
        return None
        
    return fed_tax
```

```python
print(get_federal_tax(750, 0))    # Person1: Gross 750, Code 0
print(get_federal_tax(1550, 2))   # Person2: Gross 1550, Code 2
print(get_federal_tax(1100, 5))   # Person3: Gross 1100, Code 5
```
