# Exercise 2.C Working with multiple DataFrames

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'West', 'North', 'South', 'East'],
    'Units Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
}
df_sales = pd.DataFrame(data)
```

```python
df_sales['Revenue'] = df_sales['Units Sold'] * df_sales['Price']
```

```python
total_units = df_sales['Units Sold'].sum()
total_revenue = df_sales['Revenue'].sum()
```

```python
avg_units = df_sales['Units Sold'].mean()
avg_revenue = df_sales['Revenue'].mean()
```

## Grouping Data

```python
df_sales.groupby('Product')[['Units Sold', 'Revenue']].sum()
df_sales.groupby('Region')[['Units Sold', 'Revenue']].mean()
df_sales.groupby(['Product', 'Region'])[['Units Sold', 'Revenue']].sum()
```

## Sorting and Ranking Data

```python
df_sales = df_sales.sort_values(by='Units Sold', ascending=True)
df_sales = df_sales.sort_values(by='Revenue', ascending=False)
df_sales['Revenue_Rank'] = df_sales['Revenue'].rank(ascending=False)
df_sales
```

## Merging DataFrames

```python
product_data = {'Product': ['A', 'B', 'C'], 'Category': ['Electronics', 'Furniture', 'Appliances']}
df_products = pd.DataFrame(product_data)
```

```python
df_merged = pd.merge(df_sales, df_products, on='Product')
df_merged
```

```python
df_merged.groupby('Category')['Revenue'].sum()
```
