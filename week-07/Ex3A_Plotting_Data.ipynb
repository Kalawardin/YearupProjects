# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()

plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots

```python
plt.plot(x, y, linestyle='--', color='red')
plt.show()

plt.plot(x, y, marker='s', markersize=10, markerfacecolor='green', color='green')
plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
sizes = np.random.rand(50) * 100
```

```python
plt.scatter(x_scatter, y_scatter, s=sizes, cmap='viridis', c=sizes)
plt.show()
```

## Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

x_sin = np.linspace(0, 10, 100)
y_sin = np.sin(x_sin)
ax1.plot(x_sin, y_sin)
ax1.set_title('Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('sin(X)')
```

```python
hist_data = np.random.randn(1000)
ax2.hist(hist_data, bins=30)
ax2.set_title('Histogram of Random Numbers')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
