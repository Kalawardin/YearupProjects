```python
import numpy as np
```

```python
r_1 = np.random.randint(1, 20, size=2)
print("r_1:", r_1)
```

```python
r_2 = np.random.randint(1, 20, size=(1, 2))
print("r_2:", r_2)
```

```python
r_3 = np.random.rand(2, 1)
print("r_3:\n", r_3)
```

```python
r_4 = np.random.randint(20, 100, size=(4, 5))
print("r_4:\n", r_4)
```

```python
r_5 = np.random.rand(4, 5)
print("r_5:\n", r_5)
```

```python
students = np.array(['Ali', 'Ayşe', 'Fatma', 'Ahmet', 'Mehmet', 'Can', 'Cem', 'Deniz', 'Ege', 'Eda', 'Gül', 'Naz'])
```

```python
call_on = np.random.choice(students)
print("call_on:", call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print("breakout_groups:\n", breakout_groups)
```

```python
p_values = [0.30, 0.0, 0.10, 0.10, 0.05, 0.05, 0.10, 0.05, 0.05, 0.05, 0.05, 0.10]
call_on_weighted = np.random.choice(students, size=5, p=p_values)
print("call_on_weighted:", call_on_weighted)
```

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print("bell_curve:\n", bell_curve)
```
