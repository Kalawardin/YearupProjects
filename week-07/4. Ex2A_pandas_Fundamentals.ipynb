```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}

ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))
ser4 = pd.Series(ad)
print("ser1:\n", ser1, "\nser2:\n", ser2, "\nser3:\n", ser3, "\nser4:\n", ser4)
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1, columns=['col1'])
df1

df1.insert(1, 'col2', ser2)
df1.insert(2, 'col3', ser3)
df1

df1['col4'] = ser4
df1
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
