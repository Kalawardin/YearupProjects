## Permutations

```python
print("r_4 (original):\n", r_4) 
np.random.shuffle(r_4) 
print("r_4 (shuffled):\n", r_4) 
```

```python
new_shuffle = np.random.permutation(r_4) 
print("new_shuffle:\n", new_shuffle) 
print("r_4 again:\n", r_4) 
```

## Mathematical Operations

```python
print("r_1:", r_1, r_1.shape)
print("r_2:", r_2, r_2.shape)
print("r_1 + r_2:", np.add(r_1, r_2), np.add(r_1, r_2).shape)
```

```python
print("r_1:", r_1, r_1.shape) 
print("r_3:", r_3, r_3.shape) 
print("r_1 + r_3:", np.add(r_1, r_3), np.add(r_1, r_3).shape)
```

```python
print("r_2:", r_2, r_2.shape) 
print("r_4:", r_4, r_4.shape)
```

```python
print("r_4 + [1]:\n", np.add(np.array([1]), r_4))
print("r_4 + [[1]]:\n", np.add(np.array([[1]]), r_4))
print("r_4 + [1,1,1,1,1]:\n", np.add(np.array([1,1,1,1,1]), r_4))
```

```python
print("r_3:\n", r_3) 
print("trunc:\n", np.trunc(r_3))
print("fix:\n", np.fix(r_3))
print("ceil:\n", np.ceil(r_3))
print("round:\n", np.round(r_3, 2))
```

## Cumulative Operations

```python
print("r_4:\n", r_4) 
print("Sum all:", np.sum(r_4)) 
print("Sum axis 0:", np.sum(r_4, axis=0))
print("Sum axis 1:", np.sum(r_4, axis=1)) 
print("Cumsum all:\n", np.cumsum(r_4))
print("Cumsum axis 0:\n", np.cumsum(r_4, axis=0))
```

## Grocery Problem

```python
bag1 = np.array([2, 3, 4, 0]) 
bag2 = np.array([4, 0, 1, 2]) 
print("bag1:", bag1, "bag2:", bag2)
```

```python
print(f"The first bag contains {np.sum(bag1)} items, and the second bag contains {np.sum(bag2)} items.") 
totals = bag1 + bag2
print(f"There are {totals[0]} vegetables, {totals[1]} fruits, {totals[2]} boxes of snacks, and {totals[3]} beverages in the two grocery bags.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2]) 
ben = np.array([0, 5, 5, 3, 4]) 
carla = np.array([3, 6, 0, 3, 4])
```

```python
all_knocks = np.vstack((ali, ben, carla))
print("all_knocks:\n", all_knocks)
```

```python
print("Ali total:", np.sum(ali)) 
print("Ben total:", np.sum(ben)) 
print("Carla total:", np.sum(carla)) 
print("Group total:", np.sum(all_knocks))
```

```python
daily_totals = np.sum(all_knocks, axis=0)
rolling_votes = np.cumsum(daily_totals) * 0.5
print("Rolling total estimated votes:", rolling_votes)
```
