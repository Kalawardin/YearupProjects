## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13], 
                 [14, 15, 16]])
print("arrA:\n", arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

```python
for x in arrA:
    for y in x:
        loop_result = y * 2

print("loop_result:", loop_result)
```

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print("joined_array_1:\n", joined_array_1)

joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print("joined_array_2:\n", joined_array_2)

stacked_array_1 = np.stack((joined_array_1, joined_array_1), axis=0)
print("stacked_array_1:\n", stacked_array_1)
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print("another_joined_array:\n", another_joined_array)
```

```python
stacked_h = np.hstack((arrA, arrA)) # Stacks arrays horizontally (column-wise)
stacked_v = np.vstack((arrA, arrA)) # Stacks arrays vertically (row-wise)
stacked_d = np.dstack((arrA, arrA)) # Stacks arrays depth-wise (3D)
```

```python
print("hstack:\n", stacked_h)
print("vstack:\n", stacked_v)
print("dstack:\n", stacked_d)
```

```python
splitsky = np.array_split(joined_array_1, 2)
print("splitsky:\n", splitsky)
# This results in a list of 2 separate arrays. Each array has 2 dimensions (2 rows and 3 columns)[cite: 49, 50].
```

```python
spliterino = np.array_split(joined_array_1, 3)
print("spliterino:\n", spliterino)
# This results in a list of 3 separate arrays. Since 4 rows cannot be divided evenly by 3, 
# the first array gets 2 rows, and the remaining two arrays get 1 row each. 
# All remain 2-dimensional[cite: 51, 52].
```

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print("result_1:\n", result_1)
```
